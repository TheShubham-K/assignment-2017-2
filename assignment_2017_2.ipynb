# Εύρεση Ομάδων

Ένα πρόβλημα που αντιμετωπίζουμε σε γράφους κόμβων ώστε κόμβοι που πληρούν συγκεκριμένα κριτήρια να ανήκουν στην ίδια ομάδα. Έτσι, αν έχουμε έναν γράφο που απεικονίζει ένα κοινωνικό δίκτυο, μπορεί να θέλουμε να ομαδοποιήσουμε τους κόμβους του ώστε οι κόμβοι που ανήκουν σε κάθε ομάδα να έχουν περισσότερα κοινά στοιχεία μεταξύ τους από ό,τι
με κόμβους που ανήκουν σε μια άλλη ομάδα.

Ένας τρόπος να λύσουμε το πρόβλημα αυτό είναι να διαμερίσουμε τους κόμβους του γράφου ώστε να μεγιστοποιούμε την *τμηματικότητα* (modularity) του γράφου. Η τμηματικότητα ενός γράφου ορίζεται ως εξής. Έστω ότι έχουμε δύο ομάδες $i$ και $j$. Τότε με $e_{ij}$ συμβολίζουμε το κλάσμα των συνδέσμων του γράφου που συνδέουν κόμβους της ομάδας $i$ με κόμβους της ομάδας $j$. Με την ίδια λογική, με $e_{ii}$ συμβολίζουμε το κλάσμα των συνδέσμων του γράφου που συνδέουν κόμβους της ομάδας $i$ με άλλους κόμβους της ίδιας ομάδας $i$. Επιπλέον, με $a_i$ συμβολίζουμε το κλάσμα των συνδέσμων του γράφου που καταλήγουν στην ομάδα $i$. Προσέξτε ότι αυτοί οι σύνδεσμοι περιλαμβάνουν και τους
σύνδεσμους που είναι εσωτερικοί στην ομάδα $i$ και τους συνδέσμους που ξεκινούν έξω από την ομάδα και καταλήγουν στην ομάδα. Αν σκεφτείτε λίγο, ισχύει ότι $a_i = \sum_{j} e_{ij}$. Με βάση αυτά, η τμηματικότητα είναι:

$$ Q = \sum_{i}{ e_{ii} − a_i^2} $$


Ο αλγόριθμος δουλεύει ως εξής:

1. Φτιάχνουμε μια ομάδα για κάθε κόμβο του γράφου. Άρα αν έχουμε $n$ κόμβους, τότε δημιουργούμε $n$ ομάδες του ενός κόμβου.

2. Mέχρι να φτάσουμε στον επιθυμητό αριθμό των ομάδων:
    
    2.1 Παίρνουμε κάθε ζευγάρι $i$, $j$, από τις υπάρχουσες ομάδες, όπου οι δύο ομάδες στο ζευγάρι να συνδέονται μεταξύ τους.
    
    2.2 Υπολογίζουμε τη διαφορά $2(e_{ij} − a_ia_j)$. Αποδεικνύεται ότι η διαφορά αυτή είναι ίση με τη διαφορά στην τμηματικότητα που θα προκύψει αν ενώσουμε τις δύο ομάδες $i$ και $j$: $\Delta Q = 2(e_{ij} − a_ia_j)$.
    
    2.3 Επιλέγουμε το ζευγάρι που δίνει το μεγαλύτερο $\Delta Q$ και το ενώνουμε δημιουργώντας μια νέα ομάδα.

Με τον τρόπο αυτό, δεν χρειάζεται κάθε φορά να επανυπολογίζουμε το $Q$. Αρκεί να το υπολογίσουμε στην αρχή με την αρχική κατανομή των $n$ ομάδων, και κάθε φορά που συνδέουμε δύο ομάδες θέτουμε $Q \leftarrow Q + \Delta Q$.

Έστω ότι έχουμε τον παρακάτω γράφο:

Η μέγιστη τμηματικότητα που μπορεί να προκύψει είναι $Q \approx 4{,}896$ και αντιστοιχεί στις τρεις ομάδες, κόκκινη, πράσινη και μπλε:

<img src="example_graph_1_solved.png" />

Προσέξτε ότι ο γράφος είναι μη κατευθυνόμενος, οπότε θεωρούμε ότι κάθε σύνδεσμος είναι στην πραγματικότητα διπλός (για δύο κόμβους $u$, $v$, από τον $u$ στο $v$ και από τον $v$ στο $u$). Αν η κόκκινη ομάδα είναι η $e_1$, η πράσινη ομάδα είναι η $e_2$, και η γαλάζια ομάδα είναι η $e_3$, τότε έχουμε:

$$ e_{11} = \frac{8}{24}$$
$$ e_{22} = \frac{6}{24}$$
$$ e_{33} = \frac{6}{24}$$

και:

$$ a_{1} = \frac{10}{24} $$
$$ a_{2} = \frac{7}{24} $$
$$ a_{3} = \frac{7}{24} $$

Όσον αφορά τον υπολογισμό των $e_{ij}$ προσέξτε ότι υπολογίζουμε τους συνδέσμους *από* την ομάδα $i$ στην ομάδα $j$. Έτσι, έστω ότι για παράδειγμα θέλουμε να υπολογίσουμε το $e_{ij}$ της ομάδας που προκύπτει αν συγχωνεύσουμε την ομάδα (1, 2, 3) με την ομάδα (10). Από την ομάδα (1, 2, 3) μόνο ένας σύνδεσμος υπάρχει προς την ομάδα (10), άρα $e_{ij} = 1/24$. 

Περνώντας στα $a_i$, το $a_i$ για την ομάδα (1, 2, 3) είναι το κλάσμα των συνδέσμων που καταλήγουν στην ομάδα αυτή. Συνολικά έχουμε 7 τέτοιους συνδέσμους, άρα $a_i = 7/24$. Γιατί 7 σύνδεσμοι; Διότι υπάρχουν $2 \times 3 = 6$ σύνδεσμοι που βρίσκονται μέσα στην ίδια την ομάδα (θυμηθείτε ότι αφού ο γράφος είναι μη κατευθυνόμενος, έχουμε συνδέσμους και προς τις δύο κατευθύνσεις), και ένας σύνδεσμος που βρίσκεται έξω από την ομάδα αλλά δείχνει σε αυτήν. Έτσι μπορούμε να επιβεβαιώσουμε και τον τύπο $a_i = \sum_{j} e_{ij}$ που αναφέραμε παραπάνω, αφού το $a_i$ είναι το άθροισμα του $e_{ii} = 6/24$ για την ομάδα (1, 2, 3) και του $e_{ij} = 1/24$, που μόλις είδαμε.

Όταν ενώνουμε δύο ομάδες, η διαφορά στο $Q$ θα προκύψει από το γεγονός ότι δημιουργούμε μία νέα ομάδα, η οποία περιέχει τους συνδέσμους από την ομάδα $i$ στην ομάδα $j$ και το τούμπαλιν, συνεπώς η $Q$ θα αυξηθεί κατά $e_{ij} + e_{ji}$. Θα πρέπει να αφαιρέσουμε όμως το κλάσμα των συνδέσμων προς την ομάδα $i$ *και ταυτόχρονα προς* την ομάδα $j$ (ίσο με $a_i a_j$ ) ή προς την ομάδα $j$ *και ταυτόχρονα προς* την ομάδα $i$ (ίσο με $a_j a_i$), άρα $\Delta Q = e_{ij} + e_{ji} − 2 a_i a_j$. Αλλά $e_{ij} = e_{ji}$, συνεπώς $\Delta Q = 2(e_{ij} - a_i a_j)$.

## Απαιτήσεις Προγράμματος

Κάθε φοιτητής θα εργαστεί στο προσωπικό του αποθετήριο στο GitHub. Για να αξιολογηθεί μια εργασία θα πρέπει να πληροί τις παρακάτω προϋποθέσεις:

* Όλη η εργασία θα πρέπει να βρίσκεται σε έναν κατάλογο `assignment-2017-2` μέσα στο αποθετήριο του φοιτητή.
Το πρόγραμμα θα πρέπει να έχει όνομα `community_structure.py`.

* Εννοείται ότι δεν επιτρέπεται η χρήση έτοιμων βιβλιοθηκών γράφων ή τυχόν έτοιμων υλοποιήσεων των αλγορίθμων, ή τμημάτων αυτών, πλην όσων έχουν αναφερθεί ρητώς ότι επιτρέπεται.

* Το πρόγραμμα θα μπορεί να καλείται ως εξής:
```
python community_structure.py [-n GROUPS ] graph_file
```

* Η παράμετρος `graph_file` δίνει το όνομα του αρχείου στο οποίο είναι αποθηκευμένα ο γράφος (αυτό δεν σημαίνει ότι το αρχείο ονομάζεται ντε και καλά `graph_file`, ο χρήστης μπορεί να δίνει οποιοδήποτε όνομα). Το αρχείο θα περιέχει ένα σύνδεσμο ανά γραμμή. Ο σύνδεσμος θα αφορά δύο κόμβους. Οι κόμβοι θα είναι πάντα φυσικοί αριθμοί.

* Αν δίνεται η παράμετρος `-n` τότε το `GROUPS` θα είναι ο αριθμός των ομάδων όπου θέλουμε να σταματήσουμε. Αν δεν δώσουμε την παράμετρο `-n`, τότε θα σταματήσουμε όταν φτάσουμε σε δύο ομάδες. 

Προσοχή, *η μορφή εξόδου θα πρέπει να είναι ακριβώς αυτή που περιγράφεται στη συνέχεια*.

Αν δώσουμε:
```bash
python community_structure.py -n 3 graph_file
```
θα πάρουμε στην έξοδο:
```
[1, 2, 3, 10]
[4, 5, 6]
[7, 8, 9]
Q = 0.4896
```
Όπως βλέπετε, θα υπάρχει σε κάθε γραμμή μια ομάδα, οι ομάδες θα εμφανιστούν ταξινομημένες, και μετά τις ομάδες θα εμφανιστεί η τμηματικότητα με ακρίβεια τεσσάρων δεκαδικών ψηφίων.
